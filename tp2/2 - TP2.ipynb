## Analisis de los atributos

### Atributo: group

### Atributo: still-alive             
<div>
	<ul>
		<li>El 0.757576% de los registros de este atributo, son datos faltantes.</li>

	</ul>
	Según el archivo de <a href="data/echocardiogram/echocardiogram.names">Información</a> este atributo indica con una variable binaria, si el paciente ha fallecido al final del período (0) o todavía vive (1)l Como un solo registro no tiene este atributo, decidimos borrar el registro completo. 
</div>

```python

index = d_echo_copy.index

condition = d_echo_copy["still-alive"] == "?"

faltantes_still_alive_index = index[condition]

faltantes_still_alive_index_list = faltantes_still_alive_index.tolist()


print(faltantes_still_alive_index_list)

d_echo_copy = d_echo_copy.drop(labels=faltantes_still_alive_index_list, axis=0)
d_echo_copy
```

### Atributo: survival

```python
index = d_echo_copy.index

condition = d_echo_copy["survival"] == "?"

faltantes_survival_index = index[condition]

faltantes_survival_index_list = faltantes_survival_index.tolist()


print(faltantes_still_alive_index_list)

d_echo_copy = d_echo_copy.drop(labels=faltantes_survival_index_list, axis=0)
d_echo_copy
```

### Atributo: age-at-heart-attack       
<div>
	<ul>

	</ul>
	Según el archivo de <a href="data/echocardiogram/echocardiogram.names">Información</a> <b>age-at-heart-attack</b> indica cuantos años tenía el paciente cuando sufrió el infarto. Solo seis registros no tienen este valor, asi que decidimos rellenar estos valores faltantes con la moda de la edad de todos los pacientes.

```python

moda = d_echo_copy['age-at-heart-attack'].describe()
print(moda["top"])

d_echo_copy['age-at-heart-attack']= d_echo_copy['age-at-heart-attack'].replace(to_replace="?", value=moda["top"])
d_echo_copy
```

### Atributo: fractional-shortening                  
<div>
	<ul>
		<li>El 6.060606% de los registros de este atributo, son datos faltantes.</li>
		<li>Falta 8 de los 132 datos.</li>
	</ul>
	Según el archivo de <a href="data/echocardiogram/echocardiogram.names">Información</a> <b>fractional-shortening</b> señala una medida de la contractilidad alrededor del corazón. Los números más bajos son cada vez más anormales. Ocho registros no tienen este valor.

<div>
	Vamos a ver si este atributo tiene una fuerte correlación con algún otro. Si esto se cumple, reemplazaremos esos datos faltantes teniendo en cuenta esa correlación.
</div>

```python
d_echo_copy_1 = d_echo_copy.copy(deep=False)

index = d_echo_copy_1.index

condition = d_echo_copy_1["fractional-shortening"] == "?"

faltantes_survival_index = index[condition]

faltantes_survival_index_list = faltantes_survival_index.tolist()


print(faltantes_still_alive_index_list)

d_echo_copy_1 = d_echo_copy_1.drop(labels=faltantes_survival_index_list, axis=0)


d_echo_copy_1['age-at-heart-attack'] = d_echo_copy_1['age-at-heart-attack'].astype(float)
d_echo_copy_1['still-alive'] = d_echo_copy_1['still-alive'].astype(float)
d_echo_copy_1['survival'] = d_echo_copy_1['survival'].astype(float)
d_echo_copy_1['fractional-shortening'] = d_echo_copy_1['fractional-shortening'].astype(float)

correlacion = d_echo_copy_1.corr()
correlacion
```

```python
index = d_echo_copy.index

condition = d_echo_copy["fractional-shortening"] == "?"

faltantes_fractional_shortening_index = index[condition]

faltantes_fractional_shortening_index_list = faltantes_fractional_shortening_index.tolist()


print(faltantes_fractional_shortening_index_list)

d_echo_copy = d_echo_copy.drop(labels=faltantes_fractional_shortening_index_list, axis=0)
d_echo_copy
```

### Atributo: epss             
<div>
	<ul>
		<li>El 11.363636% de los registros de este atributo, son datos faltantes.</li>
		<li>Falta 15 de los 132 datos.</li>
	</ul>
	<b>epss</b> es otra medida de contractilidad. Veremos si correlaciona con <b>fractional-shortening</b>.

```python
d_echo_copy_1 = d_echo_copy.copy(deep=False)

index = d_echo_copy_1.index

condition = d_echo_copy_1["epss"] == "?"

faltantes_epss_index = index[condition]

faltantes_epss_index_list = faltantes_epss_index.tolist()


print(faltantes_epss_index)

d_echo_copy_1 = d_echo_copy_1.drop(labels=faltantes_epss_index, axis=0)



d_echo_copy_1['fractional-shortening'] = d_echo_copy_1['fractional-shortening'].astype(float)
d_echo_copy_1['epss'] = d_echo_copy_1['epss'].astype(float)

correlacion = d_echo_copy_1.corr()
correlacion
```
