<p>En este caso, la escala es confusa y parece que las diferencias entre las categorías son mayores a los que son. </p>

<p>Si graficamos con matplotlib: </p>

```python
categories = ["Famale", "National", "Male"]
values = [0.704, 0.737, 0.783]
plt.bar(categories,values, color=["#cd8ec0","#e68800","#4d5caf"], visible="True")
for index,data in enumerate(values):
    plt.text(x=index , y =data+1 , s=f"{data}" , fontdict=dict(fontsize=20))
plt.title("GNH index by gender")
plt.show();
```

<h2>Tercer Gráfico</h2>

<p>Por último, los datos de este gráfico no están relacionados y no son excluyentes entre si. Además, no suman 100%. Se puede solucionar representados los datos con un gráfico de barras.</p>

<p>Con matplotlib: </p>

```python
import operator
from matplotlib.pyplot import figure

figure(figsize=(18, 6), dpi=80)

datos = {"WILLIAMSON": 30.23 ,
"ROHIT": 29.05 ,
"SHAKIB": 28.25 ,
"WARNER": 25.02 ,
"BABAR": 24.51 ,
"DU PLESSIS": 21.06 ,
"POORAN": 20.01 ,
"ROOT": 19.07 ,
"KUSAL PERERA": 18.16 ,
"RAHMAT SHAH": 18.16 }

colores = {
"WILLIAMSON": "#000000" ,
"ROHIT": "#03a9f5" ,
"SHAKIB": "#4eb32f" ,
"WARNER": "#f9e134" ,
"BABAR": "#409830" ,
"DU PLESSIS": "#b2eeae" ,
"POORAN": "#af2d1a" ,
"ROOT": "#8df8fe" ,
"KUSAL PERERA": "#0372fe" ,
"RAHMAT SHAH": "#4aceda" 
}

def doit(d, order):
  return  [d[k] for k in sorted(order, key=order.get, reverse=True)]

datos_ordenados = sorted(datos.items(),key=operator.itemgetter(1), reverse=True)
datos_ordenados = dict(datos_ordenados)

colores_ordenados = doit(colores,datos_ordenados)



x = datos_ordenados.keys()
y = datos_ordenados.values()

plt.bar(x,y,color=colores_ordenados)

plt.title("THE WORLD CUP'S BIG GUNS")
plt.show();
```
