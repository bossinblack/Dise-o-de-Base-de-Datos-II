## Ejercicio 1

- Descomprimir el archivo clima.tgz de la carpeta data
- Convertir cada archivo a Dataframe y mostrarlo en pantalla

```python
import pandas as pd
import tarfile
import os
import requests
```

```python

objetoTar = tarfile.open("data/clima.tgz")
objetoTar.extractall("data/clima")
objetoTar.close()


for filename in os.listdir('data/clima'):
    print(filename)
    file = 'data/clima/'+filename
    print(file)
    df = pd.read_csv(file, delimiter = "\t", encoding='utf8')
    print(df)
     
```
